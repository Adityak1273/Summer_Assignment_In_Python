```python
n=int(input("Enter the pyramid layer : "))
for i in range(0,n):
    print("*",end="")
print()    
for j in range(0,n-2):
    for k in range(0,n):
        if k==0:
            print("*",end="")
        elif k==n-1:
            print("*",end="")
        else:
            print(" ",end="")
    print()
for i in range(0,n):
    print("*",end="")    

```
