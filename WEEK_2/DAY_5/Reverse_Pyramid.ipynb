Reverse Pyramid

*********
 *******
  *****
   ***
    *
  And so on ....

```python
n=int(input("Enter the layers of pyramid : "))
add=1
n1=n
for i in range(n,0,-1):
    for k in range(1,add):
        print(" ",end="")
    if add<n:
        add+=1
    for j in range(0,1+(i-1)*2):
        print("*",end="")    
    print() 
```
