Number pyramid
    1
   121
  12321
 1234321
123454321
And so on ....

```python
n=int(input("Enter the number of pyramid layer : "))
n1=1
for i in range(n,0,-1):
    for j in range(1,i):
        print(" ",end="")
    for k in range(1,n1+1):
        print(k,end="")
    for z in range(n1-1,0,-1):
        print(z,end="")    
    if n1<n:
        n1+=1
    print()    
```
