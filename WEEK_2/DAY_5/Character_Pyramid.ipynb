Character pyramid
    A
   ABA
  ABCBA
 ABCDCBA
ABCDEDCBA
And so on ....

```python
n=int(input("Enter the number of layers in pyramid : "))
n1=1
for i in range(n,0,-1):
    for j in range(1,i):
        print(" ",end="")
    for k in range(65,65+n1):
        print(chr(k),end="")
    if n1==1:
             print(" ",end="")  
    else:
        for z in range(n1+64-1,64,-1):
            print(chr(z),end="")     
    if n1<n:
        n1+=1     
    print()    
```
