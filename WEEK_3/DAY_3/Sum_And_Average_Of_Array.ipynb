Find sum and average of array

```python
n=int(input("Enter the number of elements of an array : "))
list=[]
add=0
avg=0
for i in range(0,n):
    n1=int(input("Enter the elements of an array : "))
    list.append(n1)         
for j in list:
    add+=j
    avg=add//n
print("Sum of array : ",add)
print("Average of array : ",avg)    
```
