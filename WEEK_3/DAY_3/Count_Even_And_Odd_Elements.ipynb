Count even and odd elements

```python
n=int(input("Enter the total number of elements, you want to enter : "))
list=[]
for i in range(1,n+1):
    n1=int(input("Enter elements : "))
    list.append(n1)
even=odd=0
for i in list:
    if i%2==0:
        even+=1
    else:
        odd+=1
print("Total Even numbers : ",even)
print("Total Odd numbers : ",odd)
```
