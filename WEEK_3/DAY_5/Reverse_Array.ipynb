Reverse array
# By using not inbuilt function

```python
n=int(input("Enter the total number of elements, you want to enter : "))
list=[]
count=0
list_1=[]
for i in range(1,n+1):
    n1=int(input("Enter elements : "))
    list.append(n1)
for i in list:
    count+=1
for i in range(count-1,-1,-1):
    list_1.append(list[i])
print(list_1)
```

Reverse array
# By using inbuilt function

```python
n=int(input("Enter the total number of elements, you want to enter : "))
list=[]
count=0
for i in range(1,n+1):
    n1=int(input("Enter elements : "))
    list.append(n1)
list.reverse()
print(list)
```
