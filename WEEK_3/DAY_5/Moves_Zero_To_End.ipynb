Moves zero to end

```python
n=int(input("Enter the total number of elements, you want to enter : "))
list=[]
list_1=[]
count=0
for i in range(1,n+1):
    n1=int(input("Enter elements : "))
    list.append(n1)
print("Converting original list ",list,end="")
for i in list:
    if i==0:
        list_1.append(count)
    count+=1
for i in list_1:
    n2=0
    list.pop(i)
    list.append(n2)
print(" to zero to end list : ",list)
```
