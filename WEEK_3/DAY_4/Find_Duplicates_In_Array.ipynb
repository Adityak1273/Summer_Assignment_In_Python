Find duplicates in array
# By not using 'remove' inbuilt function 

```python
n=int(input("Enter the total number of elements, you want to enter : "))
list=[]
list_2=[]
list_3=[]
for i in range(1,n+1):
    n1=int(input("Enter elements : "))
    list.append(n1)
for i in list:
    add=0
    count=0
    list_1=[]
    for j in list:
        if j==i:
            add+=1
        if add!=1:
            list_1.append(j)
        if add==1:
            add+=1
    for k in list_1:
        if k==i:
            count+=1
        if count==1:
            list_2.append(k)
            count+=1
num1=0
for i in list_2:
    num1+=1
num2=0
for i in list_2:
    if num2<num1//2:
        list_3.append(i)
    num2+=1
print("Duplicates in Array ",list," are : ",list_3)
```

Find duplicates in array
# By using 'remove' inbuilt function 

```python
n=int(input("Enter the total number of elements, you want to enter : "))
list=[]
list_2=[]
list_3=[]
for i in range(1,n+1):
    n1=int(input("Enter elements : "))
    list.append(n1)
for i in list:
    add=0
    count=0
    list_1=[]
    list_1=list.copy()
    list_1.remove(i)
    for k in list_1:
        if k==i:
            count+=1
        if count==1:
            list_2.append(k)
            count+=1
print("Duplicates in Array ",list," are : ",list_2[:len(list_2)//2])
```
