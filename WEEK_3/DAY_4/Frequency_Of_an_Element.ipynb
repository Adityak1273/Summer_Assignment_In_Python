```python
n=int(input("Enter the total number of elements, you want to enter : "))
list=[]
count=0
for i in range(1,n+1):
    n1=int(input("Enter elements : "))
    list.append(n1)
n2=int(input("Enter the number , you want to find frequency of :"))
print("Frequency of ",n2," in array are : ",end="")
for i in list:
    if i==n2:
        count+=1
print(count)
```
