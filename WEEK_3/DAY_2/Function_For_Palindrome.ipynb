Function for palindrome

```python
def palindrome(n):
    n1=n
    rev=0
    digit=0
    while n1>0:
        digit=n1%10
        rev=rev*10+digit
        n1//=10
    if rev==n:
        print("Palindrome")
    else:
        print("Not a Palindrome")        
n=int(input("Enter the number : "))
palindrome(n)
```
