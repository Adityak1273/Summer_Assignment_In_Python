Function for perfect number

```python
def perfect(n):
    add=0
    for i in range(1,n):
        if n%i==0:
            add+=i
    if add==n:
        print(n," is a Perfect Number.")
    else:
        print(n," is not a Perfect Number.")  
n=int(input("Enter the number : "))                  
perfect(n)
```
